#Spaß mit verhältnis setzen

```python
# prompt: DataFrame „charging_sessions_df“ wird verwendet: welche UserIDs enden anders als mit ".0" oder "nan"

# Find user IDs that do not end with ".0" or "nan"
import pandas as pd
import numpy as np

# Assuming 'charging_sessions_df' is your DataFrame
# Create a boolean mask to identify users not ending with .0 or nan
mask = ~charging_sessions_df['userID'].astype(str).str.endswith(('.0', 'nan'))

# Filter the DataFrame using the mask
vorfilterung_userids_teil_eins = charging_sessions_df.loc[mask, 'userID']

# Display the filtered user IDs
vorfilterung_userids_teil_eins #schauen, ob wir das überhaupt langfristig verwenden
```

```python
# prompt: DataFrame „charging_sessions_df“ wird verwendet: stelle die spaceIDs in ein verhältnis durch ein Kreisdiagramm. also wie häufig jede spaceID jeweils verwendet wurde

#hier lässt sich sehen, dass es keine besonderen ausreißer von der reinen anzahl dre stationen gibt. eine leichte vorliebe für ca-303 ist zu erkennen
import matplotlib.pyplot as plt

# Count the occurrences of each spaceID
space_id_counts = charging_sessions_df['spaceID'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 10))  # Adjust figure size as needed
plt.pie(space_id_counts, labels=space_id_counts.index, autopct='%1.1f%%', pctdistance=.94, startangle=90, textprops={'fontsize': 8})
plt.title('Distribution of spaceIDs')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# prompt: DataFrame „charging_sessions_df“ wird verwendet: ersetze die UserIDs, die mit ".0" enden durch den leeren string "" und speichere das dataframe neu ab in mod_df

# Create a copy of the DataFrame to avoid modifying the original
mod_df = charging_sessions_df.copy()

# Replace UserIDs ending with ".0" with an empty string
mod_df['userID'] = mod_df['userID'].astype(str).str.replace(r'\.0$', '', regex=True)

#filtere im dataframe mod_df 'nan' raus
mod_df = mod_df[mod_df['userID'] != 'nan']

mod_df
```

```python
# prompt: DataFrame „mod_df“ wird verwendet: Welche Datensätze haben nicht die timezone von los angeles

# No action needed.  All rows have the same timezone value.
# To show this, you could use the following code
print(mod_df['timezone'].unique()) #hier geben mod_df und das original_df die selbe antwort
```

```python
# prompt: DataFrame „mod_df“ wird verwendet: aggregiere alle kWhdelivered einträge, die zu einer userID gehören und lege es in einem neuen dataframe ab. Sortiere danach von groß nach klein

import pandas as pd

# Group by userID and sum kWhDelivered
kwh_by_user = mod_df.groupby('userID')['kWhDelivered'].sum()

# Create a new DataFrame from the aggregated data
new_df = pd.DataFrame({'userID': kwh_by_user.index, 'total_kWhDelivered': kwh_by_user.values})

# Sort the new DataFrame by total_kWhDelivered in descending order
new_df = new_df.sort_values('total_kWhDelivered', ascending=False)

# Display the new DataFrame
new_df
```

# SpaceID analyse

```python
# prompt: DataFrame „charging_sessions_df“ wird verwendet: wie viele space IDs gibt es?

# Calculate the number of unique space IDs.
num_space_ids = charging_sessions_df['spaceID'].nunique()

# Print the result.
print(f"There are {num_space_ids} unique space IDs.")
```

```python
# prompt: ich möchte aus dem charging_sessions_df die kwhDelivered den spaceID gegenüberstellen und es grafisch abbilden

import matplotlib.pyplot as plt

# Assuming charging_sessions_df is already loaded as in the previous code

# Group by spaceID and sum kwhDelivered
kwh_by_space = charging_sessions_df.groupby('spaceID')['kWhDelivered'].sum()

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
kwh_by_space.plot(kind='bar')
plt.xlabel('Space ID')
plt.ylabel('Total kWh Delivered')
plt.title('Total kWh Delivered per Space')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# prompt: DataFrame „charging_sessions_df“ wird verwendet: sortiere die Tabelle so, dass die spaceIDs mit der höchsten kummulierten anzahl an kWhDelivered als erstes stehen

# Group by spaceID and sum kWhDelivered --> nicht mehr relevant
#space_kwh_delivered = charging_sessions_df.groupby('spaceID')['kWhDelivered'].sum()

# Sort the values in descending order
space_kwh_delivered_sorted = kwh_by_space.sort_values(ascending=False)

# Display the sorted results
space_kwh_delivered_sorted
```

```python
# Create the bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
#die 0 kann man auf 1 ändern um die erste Zeile zu ignorieren. Dann ist die grafik klarer. mglw gibts outlier, weil die spaceID auch so komisch benannt ist
space_kwh_delivered_sorted[0:].plot(kind='bar')
plt.xlabel('Space ID', fontsize=12)
plt.ylabel('Total kWh Delivered', fontsize=12)
plt.title('Total kWh Delivered per Space (Sorted)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels
plt.tight_layout()
plt.show()
```

```python
# prompt: DataFrame "space_kwh_delivered_sorted" wird verwendet: sortiere die Einträge das dataframes so, dass erst alle Einträge absteigend sortiert werden, wo die spaceID mit AG anfängt und danach all jene, dessen spaceID mit CA beginnt. Am Ende soll die spaceID 11900388 stehen.

# Assuming space_kwh_delivered_sorted is already defined as in the previous code

# Create a boolean mask for space IDs starting with 'AG'
ag_mask = space_kwh_delivered_sorted.index.str.startswith('AG')

# Create a boolean mask for space IDs starting with 'CA'
ca_mask = space_kwh_delivered_sorted.index.str.startswith('CA')

# Sort the 'AG' entries in descending order
ag_sorted = space_kwh_delivered_sorted[ag_mask].sort_values(ascending=False)

# Sort the 'CA' entries in descending order
ca_sorted = space_kwh_delivered_sorted[ca_mask].sort_values(ascending=False)

# Find the entry with spaceID 11900388
id_11900388 = space_kwh_delivered_sorted[space_kwh_delivered_sorted.index == '11900388']

# Concatenate the sorted parts and the specific entry
final_sorted = pd.concat([ag_sorted, ca_sorted, id_11900388])

# Display the final sorted results
final_sorted
```

```python
# Create the bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
#die 0 kann man auf 1 ändern um die erste Zeile zu ignorieren. Dann ist die grafik klarer. mglw gibts outlier, weil die spaceID auch so komisch benannt ist
final_sorted[0:].plot(kind='bar')
plt.xlabel('Space ID', fontsize=12)
plt.ylabel('Total kWh Delivered', fontsize=12)
plt.title('Total kWh Delivered per Space (Sorted)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels
plt.tight_layout()
plt.show()
```
